```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
```

```python


image = tf.io.read_file("spidey.jpg")

image = tf.image.decode_jpeg(image,channels=3)

image = tf.image.resize(image,(200,200))

image = image/255.

image = tf.expand_dims(image,axis=0) 
print('The shape of the image is ',image.shape)


plt.axis('off')

plt.imshow(image[0]);
```

```python
input_shape     = ( 1, 200, 200, 3)
convolved_image = tf.keras.layers.Conv2D(3, (3,3), activation='sigmoid', 
                                         input_shape=input_shape[1:])(image)
print(convolved_image.shape)

plt.imshow(convolved_image[0]);
```

```python
input_shape     = ( 1, 200, 200, 3)
convolved_image = tf.keras.layers.Conv2D(3, (5,5), 
                                         activation='relu', 
                                         kernel_initializer='glorot_uniform',
                                         input_shape=input_shape[1:])(image)
print(convolved_image.shape)
plt.imshow(convolved_image[0]);
```
